```python
# Ticket price system

customers = {
    "Alice": {"age": 15, "member": "None"},
    "Bob": {"age": 30, "member": "Silver"},
    "Charlie": {"age": 65, "member": "Gold"},
    "Diana": {"age": 8, "member": "None"}
}

totalRevenue = 0
noneCount = 0
silverCount = 0
goldCount = 0
childCount = 0
seniorCount = 0

print('===== Customer Breakdown =====')

for customer, details in customers.items():
    age = details['age']
    membership = details['member']
    basePrice = 10.00
    
    # Age Discounts
    if age < 12:
        basePrice *= 0.5
        childCount += 1
    elif age > 60:
        basePrice *= 0.7
        seniorCount += 1
    
    # Membership Discounts
    if membership == 'None':
        discountedPrice = basePrice
        noneCount += 1
    elif membership == 'Silver':
        discountedPrice = basePrice - 2
        silverCount += 1
    elif membership == 'Gold':
        discountedPrice = basePrice - 4
        goldCount += 1

    totalRevenue += discountedPrice
        
    print(f"{customer} - £{discountedPrice:.2f}")

print('\n===== Summary =====')
print(f"Total Revenue: £{totalRevenue:.2f}\n")

print("Membership Breakdown:")
print(f"None: {noneCount}")
print(f"Silver: {silverCount}")
print(f"Gold: {goldCount}\n")

print("Age Discounts Applied:")
print(f"Children: {childCount}")
print(f"Seniors: {seniorCount}")
```
