```python
employees = {
    "Sam": {"age": 28, "years_exp": 4},
    "Alex": {"age": 32, "years_exp": 7},
    "Jordan": {"age": 22, "years_exp": 1}
}

for name, details in employees.items():
    age = details["age"]
    exp = details["years_exp"]
    
    if exp < 2:
        level = "Junior"
    elif 2 <= exp <= 5:
        level = "Mid-level"
    else:
        level = "Senior"
    print(f'Name: {name}\nAge: {age}\nLevel: {level}\n')
    
    
    
    
        
```
