# ============================
#  Profile Data Structure
# ============================

profile = {
    "name": "Cameron",
    "age": 27,
    "gender": "Male",
    "town": "St Neots",
    "job": "IT Engineer",
    
    # Nested dictionary: cleaner representation of a goal + timeline
    "goal": {
        "description": "Learn AI Integration",
        "timeline_months": 12  # Adjust to 6, 12, 24, etc.
    },

    "skills": ["Quick thinking", "Hard Working"],
    "interests": ["Gym", "IT", "Animals"],
    "exp_years": 5
}

# ============================
#  Profile Output Section
# ============================

print("######## Candidate Profile #########\n")

print(f"Name: {profile['name']}")
print(f"Age: {profile['age']}  Gender: {profile['gender']}")
print(f"Location: {profile['town']}")
print(f"Job Role: {profile['job']}\n")

print("Skills:")
for skill in profile["skills"]:
    print("-", skill)

print("\nInterests:")
for interest in profile["interests"]:
    print("-", interest)

print("\nGoal:")
print("-", profile["goal"]["description"])

print("\n######## End Profile ########\n")

# ============================
#  Goal Timeline Logic
# ============================

months = profile["goal"]["timeline_months"]

if months < 12:
    goal_type = "Short-term goal"
elif months <= 24:
    goal_type = "Mid-term goal"
else:
    goal_type = "Long-term goal"

print("Goal Timeline:", goal_type)

# ============================
#  Experience Evaluation Logic
# ============================

exp = profile["exp_years"]

print("\nVerdict:")

if exp > 5:
    print("This candidate is very experienced and would be perfect for the role.")
elif exp > 3:
    print("This candidate is fairly experienced and could be suitable for the role.")
else:
    print("This candidate does not have enough experience for the role.")
